```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```
temperaturas = np.random.uniform(20, 35, 7)
print(f'As temperaturas são:  {temperaturas}')

dias = ["Seg", "Ter", "Qua", "Qui", "Sex", "Sáb", "Dom"]
print(f'Dias da semana: {dias}')
```

```
df = pd.DataFrame({"Dia ": dias, "Temperatura ": temperaturas.round(1)})
print(df)
```

```
print(f"\nMédia: {df['Temperatura '].mean(): .1f}°c")
print(f"Máxima: {df['Temperatura '].max(): .1f}°c")
print(f"Máxima: {df['Temperatura '].min(): .1f}°c")
```

```
plt.plot(dias, temperaturas, marker="o", color="orange")
plt.title("Temperatura da Semana - Linha")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura em °C")
plt.savefig("grafico_linha.png")
plt.show()
```

```
plt.bar(dias, temperaturas, color="orange")
plt.title("Temperatura da Semana - Barras")
plt.xlabel("Dias da semana")
plt.ylabel("Temeperatura em °C")
plt.savefig("grafico_barras.png")
plt.show()
```

```
plt.scatter(dias, temperaturas, color="orange", s=100)
plt.title("Temperatura da Semana - Dispersão")
plt.xlabel("Dias da semana")
plt.ylabel("Temperatura em °C")
plt.savefig("grafico_dispersao.png")
plt.show()
```
